```python
fact=1
n= int(input('Enter the number:'))
for i in range(1,n+1):
    fact=fact*i
print('The factorial is:',fact)
```

```python
n= int(input('Enter the number:'))
c=0
for i in range(1,n+1):
    if n%i==0:
        c=c+1
if c==2:
    print('Prime Number')
else:
    print('Not Prime ')
```

```python
n= int(input('Enter the terms:'))
a=0
b=1
print(a,b,end=" ")
for i in range(2,n):
    c=a+b
    print(c,end=' ')
    a=b
    b=c
```
