```python
class Hello:
    def run(self):
        for i in range(5):
            print('Hello')
    
    

class Hi:
    def run(self):
        for i in range(5):
            print('Hi')
            
t1 = Hello()
t2 = Hi()

t1.run()
t2.run()
```

```python
from threading import *
```

```python
class Hello(Thread):
    def run(self):
        for i in range(5):
            print('Hello')
    
    

class Hi(Thread):
    def run(self):
        for i in range(5):
            print('Hi')
            
t1 = Hello()
t2 = Hi()

t1.run()
t2.run()
```

```python
class Hello(Thread):
    def run(self):
        for i in range(5):
            print('Hello')
    
    

class Hi(Thread):
    def run(self):
        for i in range(5):
            print('Hi')
            
t1 = Hello()
t2 = Hi()

t1.start()
t2.run()
```

```python
from time import sleep
```

```python
class Hello(Thread):
    def run(self):
        for i in range(500):
            print('Hello')
            sleep(1)
    
    

class Hi(Thread):
    def run(self):
        for i in range(500):
            print('Hi')
            sleep(1)
            
t1 = Hello()
t2 = Hi()

t1.start()
t2.start()
```

```python
class Hello(Thread):
    def run(self):
        for i in range(50):
            print('Hello')
            sleep(1)
    
    

class Hi(Thread):
    def run(self):
        for i in range(50):
            print('Hi')
            sleep(1)
            
t1 = Hello()
t2 = Hi()

t1.start()
sleep(0.2)
t2.start()
```
