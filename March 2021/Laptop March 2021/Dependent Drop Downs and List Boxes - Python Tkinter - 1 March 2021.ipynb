```python
from tkinter import *
from tkinter import ttk

window = Tk()
window.title('Open Dialog box')
window.geometry('500x400')
window.iconbitmap(r'C:\Users\ARVIND\Desktop\Desk\Bevy\av.ico')

# creating a list of sizes
size = ['small','medium','large']

small_colors = ['red','green','Blue','Black']
medium_colors = ['red','green']
large_colors = ['Blue','Black']

def pick_color(e):
	if my_combo.get() == 'small':
		color_combo.config(value=small_colors)
		color_combo.current(0)
	if my_combo.get() == 'medium':
		color_combo.config(value=medium_colors)
		color_combo.current(0)
	if my_combo.get() == 'large':
		color_combo.config(value=large_colors)
		color_combo.current(0)

# create a drop down box

my_combo = ttk.Combobox(window,value=size)
my_combo.current(0)
my_combo.pack(pady=20)

# bind the combo box

my_combo.bind('<<ComboboxSelected>>',pick_color)

color_combo = ttk.Combobox(window,value=[' '])
color_combo.current(0)
color_combo.pack(pady=20)


#Frame

my_frame = Frame(window)
my_frame.pack(pady=50)

# List Boxes
my_list1 = Listbox(my_frame)
my_list2 = Listbox(my_frame)
my_list1.grid(row=0,column=0)
my_list2.grid(row=0,column=1,padx=20)

def list_color(e):

	if my_list1.get(ANCHOR) == 'small':
			for item in small_colors:
				my_list2.insert(END,item)

	if my_list1.get(ANCHOR) == 'medium':
			for item in medium_colors:
				my_list2.insert(END,item)

	if my_list1.get(ANCHOR) == 'large':
			for item in large_colors:
				my_list2.insert(END,item)

# add items to list1

for item in size:
	my_list1.insert(END,item)

# Binding the listbox

my_list1.bind('<<ListboxSelect>>',list_color)

window.mainloop()
```

# Delete items from Listbox

```python
from tkinter import *
from tkinter import ttk
window = Tk()
window.title('List box')
window.geometry('500x400')
window.iconbitmap(r'C:\Users\ARVIND\Desktop\Desk\Bevy\av.ico')
# Listbox
my_listbox = Listbox(window)
my_listbox.pack(pady=20)
# add items to listbox
my_listbox.insert(END,'This is an item')
my_listbox.insert(END,'Second item')

# add list of items
my_list = ['one','Two','Three']

for item in my_list:
	my_listbox.insert(END,item)

def delete():
	my_listbox.delete(ANCHOR)

my_button = Button(window,text='Delete',command=delete)
my_button.pack(pady=10)
window.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
window = Tk()
window.title('List box')
window.geometry('500x400')
window.iconbitmap(r'C:\Users\ARVIND\Desktop\Desk\Bevy\av.ico')
my_listbox = Listbox(window)
my_listbox.pack(pady=20)
my_listbox.insert(END,'This is an item')
my_listbox.insert(END,'Second item')
my_list = ['one','Two','Three']

for item in my_list:
	my_listbox.insert(END,item)

def delete():
	my_listbox.delete(ANCHOR)
	my_label.config(text=' ')

def select():
	my_label.config(text=my_listbox.get(ANCHOR))

my_button = Button(window,text='Delete',command=delete)
my_button.pack(pady=10)

my_button2= Button(window,text='Select',command=select)
my_button2.pack(pady=10)

global my_label
my_label = Label(window,text=' ')
my_label.pack(pady=5)

window.mainloop()
```
